## Ecuaciones de Volterra-Lotka

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def Presa_Depredador(h:float,t:int,x:int,y:int):
    """La función Presa_Depredador modela el comportamiento de las ecuaciones de Lodka-Volterra:
            x'(t) = ax(t) -bx(t)y(t)
            y'(t) =-cy(t) +dx(t)y(t) 
    representan el numero de presas y depredadores a lo largo de un periodo de tiempo(t).

    Args:
        h (float): Incrementos durante el tiempo
        t (int): Tiempo(días)
        x (int): Numero inicial de presas en miles con t=0
        y (int): Numero inicial de depredadores en miles con t=0
    """
    X=[]
    Y=[]
    # Tasa de crecimiento de las presas sin depredadores.
    a = 0.4 
    # Capacidad de supervivencia de las presas, interación presa-depredador. 
    b = 0.3
    # Tasa de decrecimiento de los depredadores sin presas.
    c = 0.37
    # Capacidad de depredación de los depredadores, interación (depredador-presa).
    d = 0.05 
    
    X.append(round(x,4)),Y.append(round(y,4))
    h_0 = 0 

    #Metodo de Euler para resolver el sistema de Ecuaciones Diferenciales 
    for i in range(0,t):
        f1 = (a*x)-(b*x*y)
        f2 = (-c*y)+(d*x*y)
        x = round(x + (h*f1),4)
        y = round(y + (h*f2),4)
        X.append(x)
        Y.append(y)
        h_0 = h_0 + h
        
    tabla = pd.DataFrame(list(zip(X,Y)),columns=['Presas','Depredadores'])
    
    print(tabla)
    tabla.plot(figsize=(12,6),xlabel='Tiempo (días)',ylabel='Presas/Depredadores (miles)')
```

##### Pruebas 

```python
#Numero de presas > Numero de depredadores 
h = 0.01 
t = 10000
x = 3
y = 1

Presa_Depredador(h,t,x,y)
```

```python
#Numero de presas < Numero de depredadores 

h = 0.01 
t = 10000
x = 1
y = 3

Presa_Depredador(h,t,x,y)
```

```python
# Numero de presas = Numero de depredadores 
h = 0.01 
t = 10000
x = 3
y = 3

Presa_Depredador(h,t,x,y)
```
